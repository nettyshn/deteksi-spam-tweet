# LOAD DATA

```python
# Setelah kita mengambil data (crawling data) melalui Twitter API pada tahapan sebelumnya dan telah disimpan dalam file .csv,
# maka kemudian kita load data tersebut untuk tahapan preprocessing data yang bertujuan untuk mempersiapkan teks tersebut
# menjadi data yang dapat digunakan pada proses berikutnya yaitu proses klasifikasi spam dengan metode Naive Bayes.

def load_data():
    spamtweet = pd.read_csv('dataset_kotor.csv') # diasumsikan bahwa dataset berada pada folder tersendiri.
    return spamtweet
```

```python
# load data dalam bentuk dataframe agar terlihat lebih tertata dan mudah dimengerti
tweet_df = load_data()
tweet_df.head()
```

```python
# untuk mengetahui ukuran dataset dan column dari data yang kita miliki
print('Dataset size:',tweet_df.shape)
print('Columns are:',tweet_df.columns)
```

# PREPROCESSING DATA

## Text Cleaning

```python
# Tahap ini bertujuan untuk menghilangkan tanda baca ataupun karakter selain teks.
# Tahap ini biasa juga disebut dengan punctuation removal.

string.punctuation
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

tweet_df['Tweet_Punct'] = tweet_df['Tweets'].apply(lambda x: remove_punct(x))
tweet_df.head(10)
```

## Case Folding and Tokenizing

```python
# pada baris ini, kami menggunakan code dengan menggabungkan 2 tahapan yaitu case folding dan tokenizing, dimana
# Case Folding merupakan tahapan yang bertujuan untuk mengubah semua huruf di dalam text menjadi huruf kecil, sedangkan
# Tokenizing merupakan tahapan yang bertujuan untuk melakukan proses pemisahan setiap kata yang disebut dengan token.

def tokenization(text):
    text = re.split('\W+', text)
    return text

tweet_df['Tweet_Tokenized'] = tweet_df['Tweet_Punct'].apply(lambda x: tokenization(x.lower()))
tweet_df.head(10)
```

## Stopword Removal

```python
# Tahap ini bertujuan untuk menghapus kaa-kata yang terlalu umum dan kurang penting,
# seperti banyaknya kemunculan suatu kata yang berulang-ulang atau jumlahnya cukup
# banyak dibandingkan dengan kata lainnya.

stopword = nltk.corpus.stopwords.words('indonesian')
#list_stopwords = stopwords.words('indonesian')
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
tweet_df['Tweet_Stopwords'] = tweet_df['Tweet_Tokenized'].apply(lambda x: remove_stopwords(x))
tweet_df.head(10)
```

## Lemmatization

```python
# pada tahap ini bertujuan untuk mengenali kata-kata yang memiliki bentuk yang beda tetapi
# sebenarnya adalah kata yang sama.

wn = nltk.WordNetLemmatizer() #disini kami melakukan lemmatization dengan WordNetLemmatizer sesuai yang telah diajarkan di matakuliah PBA
                                #karena dalam NLTK wordnet adalah yang paling kuat dalam lemmatization.

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

tweet_df['Tweet_Lemmatized'] = tweet_df['Tweet_Stopwords'].apply(lambda x: lemmatizer(x))
tweet_df.head(10)

    
```
